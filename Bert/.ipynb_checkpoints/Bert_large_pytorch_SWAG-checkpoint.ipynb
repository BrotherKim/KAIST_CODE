```python
SWAG_DIR = '/content/pytorch_pretrained_BERT/examples/swagaf'
OUTPUT_DIR = '.'

!python run_swag.py \
  --bert_model bert-large-uncased \
  --do_train \
  --do_lower_case \
  --do_eval \
  --data_dir $SWAG_DIR/data \
  --train_batch_size 16 \
  --learning_rate 2e-1 \
  --num_train_epochs 3.0 \
  --max_seq_length 80 \
  --output_dir $OUTPUT_DIR/swag_output/ \
  --gradient_accumulation_steps 4
```
