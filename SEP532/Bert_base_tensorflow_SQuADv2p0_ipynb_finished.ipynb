### **Set up your TPU environment**
*   Verify that you are connected to a TPU device
*   You will get know your TPU Address that is used at time of fine-tuning
*   Perform Google Authentication to access your bucket
*   Upload your credentials to TPU to access your GCS bucket

```python
import datetime
import json
import os
import pprint
import random
import string
import sys
import tensorflow as tf

assert 'COLAB_TPU_ADDR' in os.environ, 'ERROR: Not connected to a TPU runtime; please see the first cell in this notebook for instructions!'
TPU_ADDRESS = 'grpc://' + os.environ['COLAB_TPU_ADDR']
#print('TPU address is => ', TPU_ADDRESS)

from google.colab import auth
auth.authenticate_user()
with tf.Session(TPU_ADDRESS) as session:
  print('TPU devices:')
  pprint.pprint(session.list_devices())

  # Upload credentials to TPU.
  with open('/content/adc.json', 'r') as f:
    auth_info = json.load(f)
  tf.contrib.cloud.configure_gcs(session, credentials=auth_info)
  # Now credentials are set for all future sessions on this TPU.

os.environ['TPU_ADDRESS'] = TPU_ADDRESS
print('TPU address is => ', os.environ['TPU_ADDRESS'])
```

### **Create output directory** 


> Need to create a output directory at GCS (Google Cloud Storage) bucket, where you will get your fine_tuned model after training completion. For that you need to provide your BUCKET name and OUPUT DIRECTORY name.

> Also need to move Pre-trained Model at GCS (Google Cloud Storage) bucket, as Local File System is not Supported on TPU. If you don't move your pretrained model to TPU you may face an error. 




```python
BUCKET = 'bk_bert_squad_v2p0' #@param {type:"string"}
assert BUCKET, '*** Must specify an existing GCS bucket name ***'
output_dir_name = os.environ['BASE_MODEL'] + '_out' #@param {type:"string"}

os.environ['BUCKET_NAME'] = 'gs://{}'.format(BUCKET)
os.environ['OUTPUT_DIR'] = 'gs://{}/{}'.format(BUCKET,output_dir_name)
tf.gfile.MakeDirs(os.environ['OUTPUT_DIR'])
print('***** Model output directory: {} *****'.format(os.environ['OUTPUT_DIR']))
```

### **Move Pretrained Model to GCS Bucket** 


> Need to move Pre-trained Model at GCS (Google Cloud Storage) bucket, as Local File System is not Supported on TPU. If you don't move your pretrained model to TPU you may face the error. 



> The **gsutil** **mv** command allows you to move data between your local file system and the cloud, move data within the cloud, and move data between cloud storage providers.




```python
!gsutil mv /content/bert/$BASE_MODEL $BUCKET_NAME
```

### **Training**

> Set options for TPU.

1.   --use_tpu=True
2.   --tpu_name="find with 'TPU address is =>"





```python
!python run_squad.py \
  --vocab_file=$BUCKET_NAME/$BASE_MODEL/vocab.txt \
  --bert_config_file=$BUCKET_NAME/$BASE_MODEL/bert_config.json \
  --init_checkpoint=$BUCKET_NAME/$BASE_MODEL/bert_model.ckpt \
  --do_train=True \
  --train_file=train-$SQUAD_VERSION.json \
  --do_predict=True \
  --predict_file=dev-$SQUAD_VERSION.json \
  --train_batch_size=24 \
  --learning_rate=3e-5 \
  --num_train_epochs=2.0 \
  --use_tpu=True \
  --tpu_name=$TPU_ADDRESS \
  --max_seq_length=384 \
  --doc_stride=128 \
  --version_2_with_negative=True \
  --output_dir=$OUTPUT_DIR 
```
